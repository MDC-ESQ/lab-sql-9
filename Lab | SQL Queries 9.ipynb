1. How many distinct (different) actors' last names are there?

```python

data1 = pd.read_sql_query('SELECT * FROM actor', engine)

query1 = '''SELECT COUNT(DISTINCT last_name)
FROM sakila.actor'''
print(query1)
data1 = pd.read_sql_query(query1, engine)
data1.head()
```

2. Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.

```python

data2 = pd.read_sql_query('SELECT * FROM rental', engine)

query2 = '''SELECT 
    *,
    CASE
	WHEN DAYNAME(rental_date) = 'Saturday' or DAYNAME(rental_date) = 'Sunday' then 'weekend'
	ELSE 'workday'
	END AS 'day_type'
FROM
    sakila.rental'''
print(query2)
data2 = pd.read_sql_query(query2, engine)
data2.head()
```

3. Get all films with ARMAGEDDON in the title.

```python
data3 = pd.read_sql_query('SELECT * FROM film', engine)

query3 = '''SELECT * FROM sakila.film
WHERE title LIKE "%%ARMAGEDDON%%"'''
print(query3)
data3 = pd.read_sql_query(query3, engine)
data3.head()
```

4. Get 10 the longest films.

```python
data4 = pd.read_sql_query('SELECT * FROM film', engine)

query4 = '''SELECT * FROM sakila.film
ORDER BY length DESC
LIMIT 10'''
print(query4)
data4 = pd.read_sql_query(query4, engine)
data4.head()
```

5. How many films include Behind the Scenes content?

```python
data5 = pd.read_sql_query('SELECT * FROM film', engine)

query5 = '''SELECT COUNT(film_id) FROM sakila.film
WHERE special_features = "Behind the Scenes"'''
print(query5)
data5 = pd.read_sql_query(query5, engine)
data5.head()
```

6. Which kind of movies (rating) have a mean duration of more than two hours?

```python
data6 = pd.read_sql_query('SELECT * FROM film', engine)

query6 = '''SELECT c.name, AVG(f.length) as 'Average duration'
FROM sakila.category c
INNER JOIN sakila.film_category f_c USING (category_id)
INNER JOIN sakila.film f USING (film_id)
GROUP BY c.name
HAVING AVG(f.length) > 120'''
print(query6)
data6 = pd.read_sql_query(query6, engine)
data6.head()
```

7. 
Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank.

```python
data7 = pd.read_sql_query('SELECT * FROM film', engine)

query7 = '''SELECT title, length, DENSE_RANK() OVER (ORDER BY length DESC) AS 'rank'
FROM sakila.film'''
print(query7)
data7 = pd.read_sql_query(query7, engine)
data7.head(20)
```
